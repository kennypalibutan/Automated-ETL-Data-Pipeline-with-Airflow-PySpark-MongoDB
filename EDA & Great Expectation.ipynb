### Great Expectation

````python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
````

````python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'amazon_cleaned_datasource7'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'amazon_cleaned_asset'
path_to_data = 'amazon_cleaned.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
````

````python
# Creat an expectation suite
expectation_suite_name = 'suite1' #menamakan suite
context.add_or_update_expectation_suite(expectation_suite_name) #membuat suite

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
````

### Expectation 1

````python
validator.expect_column_values_to_be_unique(column="OrderID")
````

### Expectation 2

````python
validator.expect_column_values_to_be_between(
    column="Discount",
    min_value=0,
    max_value=1.0
)
````

### Expectation 3

````python
validator.expect_column_values_to_be_in_set(
    column="OrderStatus",
    value_set=["Delivered", "Cancelled", "Pending", "Shipped", "Returned"],
    mostly=1.0  # Semua baris harus sesuai
)
````

### Expectation 4

````python
validator.expect_column_values_to_be_in_type_list(
    column="Quantity",
    type_list=["int64"]
)
````

### Expectation 5

````python
validator.expect_column_values_to_not_be_null(
    column="ProductName"
)
````

### Expectation 6

````python
validator.expect_column_pair_values_a_to_be_greater_than_b(
    column_A="UnitPrice",
    column_B="Tax"
)
````

### Expectation 7

````python
validator.expect_column_to_exist(
    column="OrderDate"
)
````
